```python
# Oppgave 4)

# Finn skjæringspunktene mellom f(x) = x^3 - 2x + 1 og g(x) = e^x

# Først kan vi si at f(x) = g(x). Så kan vi omforme en ny h(x) som er gitt ved nullpunktet til den nye funksjonen.

# f(x) = g(x), x^3 - 2x + 1 = e^x, h(x) = x^3 - 2x + 1 - e^x. Skjæringspunktet er ved h(x) = 0. 

# Noe annet er å sjekke for flere skjæringspunkter. Her kan vi plotte funksjonene først for å se hvor de skjærer. 

from pylab import * 

def f(x):
    return x**3 - 2*x + 1

def g(x):
    return e**x

# Plotter funksjonene
x = linspace(-3, 4.5, 100)
plot(x, f(x))
plot(x, g(x))
show()

# Vi ser at det er flere nullpunkter.

#Bruker Newtons metode for å finne nullpunktene. 

def h(x):
    return x**3 - 2*x + 1 - e**x

def hder(x):
    return 3*x**2 - 2 - e**x 

# Newtons metode

def NewtonsMetode(f, fder, x, tol=1E-5):
    while abs(f(x)) > tol:
        x = x - f(x)/fder(x)
    return x

x1 = NewtonsMetode(h, hder, 4)
x2 = NewtonsMetode(h, hder, 2)
x3 = NewtonsMetode(h, hder, -0.5)
x4 = NewtonsMetode(h, hder, -3)

print("Skjæringspunktenes x verdier er", x1, x2, x3, x4)

s1 = (x1, f(x1))
s2 = (x2, f(x2))
s3 = (x3, f(x3))
s4 = (x4, f(x4))

print("Skjæringspunktene er", s1, s2, s3, s4)
```

```python
# Oppgave 5) 
from pylab import * 

# Skriver opp listene

tid = [0,15,25,45,60,90]
blodsukker = [5.2,6.5,7.8,6.5,5.4,5.3]
endring = zeros(N)

N = len(tid)

# Deriverte 
for i in range(N-1):
    db = blodsukker[i+1] - blodsukker[i]
    dt = tid[i+1] - tid[i]
    endring[i] = db/dt
    
# Plotter funksjonene

plot(tid, blodsukker)
plot(tid, endring)
show()
```

```python
# Oppgave 6) 

# Her kan vi bruke differensiallikninger, Eulers metode.

# B'(t) = k*B(t)

from pylab import * 

b0 = 1000
k = 0.62 

dt = 1E-5
tid = 15 #timer
N = int(tid/dt)

t = zeros(N+1)
b = zeros(N+1)

def bder(B):
    return k*B

b[0] = b0

# Eulers metode

for i in range(N):
    b[i + 1] = b[i] + bder(b[i])*dt
    t[i + 1] = t[i] + dt
    
plot(t, b)
show()

```
