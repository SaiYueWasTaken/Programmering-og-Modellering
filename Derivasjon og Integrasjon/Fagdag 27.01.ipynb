```python
#Oppgave 7.3

from pylab import * 

def derivert(f, x, delta_x):
    fder = (f(x + delta_x) - f(x))/delta_x
    return fder

def N(x):
    return 0.2*e**x + 90
t_start = 0
t_slutt = 10
t = linspace(t_start, t_slutt, 10000)
dt = 1E-9

N_antall = N(t)
N_der = derivert(N, t, dt)

plot(t, N_der, color = "green", label = "Stigningen")
plot(t, N_antall, color ="navy", label = "Antall")
legend()
xlabel("Tid (år)")
ylabel("Antall Planter")
show()
```

```python
#Oppgave 7.14

from pylab import *

def f(x):
    return 2*x + 1 

#Integralet av 2x + 1 er x^2 + x + C 

a = 1 
b = 4 

#Det bestemte integralet vil da bli 18

A = 0
#N = 3 
N = 100
def rektangelmetoden(f, a, b, n):
    total = 0
    dx = (b-a)/n
    for k in range(0,n):
        total += f(a + (k*dx))
    return dx*total

def trapesmetoden(f, a, b, n):
    dx = (b-a)/n
    total = (f(a) + f(b))/2
    for k in range(1,n):
        total += f(a + k*dx)
    return dx*total


print(f"Ved bruk av rektangelmetoden er Arealet {rektangelmetoden(f, a, b, N)}")

print(f"Ved bruk av trapesmetoden er Arealet {trapesmetoden(f, a, b, N)}")


#Vi ser at ved ulike verdier av N vil rektangelmetoden komme nærmere det trapesmetoden får. 
#Siden trapesmetoden får kantene også under funksjonen f(x) sitt Areal (Integralet) vil 
#den være mer riktig. Dette kan vi se ved at den vil gi 18 som svar, som er likt det jeg 
#regnet ut.
#Rektangelmetoden derimot vil ha et mindre riktig svar når det er færre rektangler, men det
#tilnærmer seg 18 ved at når rektanglene blir så liten at de går mot 0, vil kantene også
#bli mye mindre. Dette gjør at Rektangelet blir nærmere etter som N blir større.
```

Teoridel: Analytisk og numerisk

#Derivasjon: 

Når man snakker om analytisk derivasjon snakker man om matte måten å regne derivasjon på. Da kan vi feks regne ut den deriverte av et utrykk. Dette kan da være:

f(x) = 2x^2 + 1 
f'(x) = 4x 

Hvis vi setter f'(2), får vi 8.  

Hvis vi snakker numerisk, er det sånn at vi kan variere delta_x. Endringen av x gitt ved endring av f(x). Dette kan gi oss stigningen til funskjonen. I analytisk derivasjon setter vi en limit at dx -> uendelig. Men det gjør vi litt anneledes i programmering. I programmering kan vi bare sette inn et "stort tall" noe som vi tilnærmer uendelig. Vi tilnærmer da Newtons kvotient. 

#Integrasjon:

I integrasjon handler det å finne ut arealet under en funksjon. Ved hjelp av integrasjon kan vi finne et funksjonsutrykk og videre igjen finne arealet. 

I analytisk integrasjon regner vi oss ut ved bestemte integraler og igjennom funksjoner og funksjonsverdier. Dette kan vi gjøre ved hjelp av:

f(x)= 2x + 1 
F(x) = x^2 + x + C (hvor C er en konstant) 

Ved dette setter vi opp ett bestemt integral (noe som gjør at C forsvinner). 

I numerisk integrasjon handler det om størrelsen på antall rektangler. Dette via Rektangelmetoden vi har lært. Det er flere metoder å regne Arealet på, men den vi har kommet mest inn på er rektangelmetoden. Denne metoden setter opp rektangler som kan simulere et areal under en funksjon. Dette arealet er ikke helt perfekt i forhold til arealet under grafen, siden det kommer mellomrom (trekanter) mellom funksjonen og rektangelene. Ved å justere N, kan vi endre antall rektangler, altså hvor brede rektangelen er, slik at vi kan gjøre størrelsen på disse trekantene infinidesimalt (hehe), noe som tilnærmer det arealet vi kan regne ved analytisk.
