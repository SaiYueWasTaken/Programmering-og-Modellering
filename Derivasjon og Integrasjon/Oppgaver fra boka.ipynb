# Integrasjons Oppgaver 


```python
# Oppgave 7.14

def f(x):
    return 2*x + 1

def f2(x):
    return x**2 - 4*x + 5

def f3(x):
    return e**x

def f4(x):
    return e**x - 5*x

def rektangelmetoden(f, a, b, n):
    A = 0
    h = (b-a)/n
    for i in range(n):
        A = A + f(a + i*h)
    areal = A*h
    return areal

def trapesmetoden(f, a, b, n):
    A = 0
    h = (b-a)/n
    for i in range(n):
        A = A + f(a + i*h)
    areal = h*(A + (f(a) + f(b))/2) 
    return areal

Integral1_rekt = rektangelmetoden(f, 1, 4, 1000)
Integral1_trap = trapesmetoden(f, 1, 4, 1000)
print(Integral1_rekt, Integral1_trap)
```
