# Vann i tank; Difflikning

## $$y'(t) = -0.25 * y(t)$$

```python
from pylab import *

#Initialbetingelser
a = 0
b = 10
dt = 1E-5
N = int((b-a)/dt)

x = zeros(N+1)
y = zeros(N+1)
y[0] = 6000
x[0] = 0

def fder(x, y):
    return -0.25*y

for n in range(N):
    y[n+1] = y[n] + fder(x[n], y[n])*dt
    x[n+1] = x[n] + dt

plot(x,y)
grid()
show()
```

Ved å bruke Eulers metode har jeg løst denne difflikningen. Utformingen av Eulers metode ser man i "Differensiallikninger - Fagdag 17.03.2021".
