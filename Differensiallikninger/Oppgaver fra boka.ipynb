```python
# Oppgave 8.2) 
from pylab import *

x_start = 0
x_slutt = 10
dx = 1E-5 
N = int((x_slutt-x_start)/dx)
y0 = 0

def yder(y, x):
    return 1 

def yder2(y, x):
    return x

def yder3(y, x):
    return 4 + 3*y 

def yder4(y, x):
    return -4*y - 3*x

def yder5(y, x):
    return 3*x + 4 - y/x

def yder6(y, x):
    return x + y**2

x = zeros(N + 1) # +1, fordi det er et ekstra tidssteg, så vi trenger en ekstra array. 
y = zeros(N + 1)
y[0] = y0

#Eulers metode
for i in range(N):
    y[i + 1] = y[i] + yder(y[i], x[i])*dx
    x[i + 1] = x[i] + dx 
    

plot(x,y)
show()
    
```

```python
# Oppgave 8.16) 
# a)
from pylab import * 

# Newtons avkjølingslov; T'(t) = -k(T(t) - T_0)

t_start = 0
t_slutt = 30
N = 10000
dt = (t_slutt - t_start)/(N - 1)
t0 = 7

def Tder(T):
    return -0.16*(T - 20)


t = zeros(N)
T = zeros(N)
T[0] = t0

# Eulers metode
for i in range(N-1):
    T[i + 1] = T[i] + Tder(T[i])*dt
    t[i + 1] = t[i] + dt
    
# b)

plot(t, T)
show()



   
```

```python
# c) 

from pylab import * 

# Newtons avkjølingslov; T'(t) = -k(T(t) - T_0)

t_start = 0
t_slutt = 30
N = 10000
dt = (t_slutt - t_start)/(N - 1)
t0 = 7


def Tder(T, t):
    return -0.16*(T - ((5*cos((pi**2)*t/15)) + 19))

t = zeros(N)
T = zeros(N)
T[0] = t0

# Eulers metode
for i in range(N-1):
    T[i + 1] = T[i] + Tder(T[i], t[i])*dt
    t[i + 1] = t[i] + dt

plot(t, T)
show()

```
